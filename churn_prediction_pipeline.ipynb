```python
# Import libraries

import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# load dataset

df=pd.read_csv('/content/customer_churn.csv')
df
```

```python
# handling missing value

# Convert TotalCharges to numeric (common issue)
if "TotalCharges" in df1.columns:
    df1["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
# fill numeric columns

num_cols = df1.select_dtypes(include=["int64", "float64"]).columns
df1[num_cols] = df1[num_cols].fillna(df[num_cols].median())
```

```python
# Fill categorical columns

cat_cols = df1.select_dtypes(include=["object"]).columns
df1[cat_cols] = df1[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

```python
df1
```

```python
df1.dtypes
```

```python
# encodimg categorical values

label_encoder = LabelEncoder()
df1['Churn'] = label_encoder.fit_transform(df1['Churn'])
```

```python
# Binary encoding

binary_cols = ["PaperlessBilling"]
for col in binary_cols:
    df1[col] = df1[col].replace({"Yes": 1, "No": 0})
```

```python
# one hot encoding - nominal

df1 = pd.get_dummies(
    df1,
    columns=["Contract", "PaymentMethod"],
    drop_first=True
)
```

```python
df1
```

```python
# feature scaling

# Minmax scaling
minmax = MinMaxScaler()
df1[["MonthlyCharges"]] = minmax.fit_transform(df1[["MonthlyCharges"]])
```

```python
# standard scaling

standard = StandardScaler()
df1[["Tenure"]] = standard.fit_transform(df1[["Tenure"]])
```

```python
df1
```

```python
# Outlier detection

#IQR method
Q1 = df1["MonthlyCharges"].quantile(0.25)
Q3 = df1["MonthlyCharges"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df1  = df1[(df1["MonthlyCharges"] >= lower) & (df1["MonthlyCharges"] <= upper)]
```

```python
# feature engineering

# 1. Customer Lifetime Value
df1["Customer_Lifetime_Value"] = df1["MonthlyCharges"] * df1["Tenure"]
```

```python
#  2. High Value Customer
df1["High_Value_Customer"] = (
    df1["Customer_Lifetime_Value"] > df1["Customer_Lifetime_Value"].median()
).astype(int)
```

```python
# 3. Tenure Group
df1["Tenure_Group"] = pd.cut(
    df1["Tenure"],
    bins=[-2, -0.5, 0.5, 2],
    labels=["New", "Medium", "Long"]
)
```

```python
# 4. Payment Efficiency
df1["Payment_Efficiency"] = df1["TotalCharges"] / (df1["Tenure"] + 1)
```

```python
# 5. High Monthly Charge Flag
df1["High_Monthly_Charge"] = (
    df1["MonthlyCharges"] > df1["MonthlyCharges"].mean()
).astype(int)
```
